```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()

# Display the shapes of the datasets
print("First 5 samples of training data: \n", train_data[:5])
print("First 5 training targets: \n", train_targets[:5])
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize the data (mean=0, variance=1)

scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)
```

```python
# Build the model
import tensorflow as tf
from tensorflow.keras import models, layers
from sklearn.preprocessing import StandardScaler
# Build the model
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(train_data.shape[1],)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1))  # Single output for regression

# Compile the model
model.compile(optimizer='Adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(train_data, train_targets, epochs=100)

# Evaluate the model on the test data
test_mse_score, test_mae_score = model.evaluate(test_data, test_targets)

print(f"Mean Absolute Error on test data: {test_mae_score}")
```

```python
import numpy as np
baseline_mae = np.mean(np.abs(test_targets - np.mean(train_targets)))
print(f"Baseline MAE: {baseline_mae}")
```
